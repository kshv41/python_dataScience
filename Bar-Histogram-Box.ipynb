# Analysing movie ratings on different websites 

```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import arange
```

```python
ratings = pd.read_csv('fandango_scores.csv')
```

## Bar plots

```python
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

fig, ax = plt.subplots()
position = arange(5)+ 0.75
height = ratings.loc[:, num_cols ].iloc[0, :]
width = 0.5
ax.bar(position, height, width)
ax.set_xticks(arange(5)+ 0.75)
ax.set_xticklabels(num_cols, rotation = 90)
ax.set_xlabel('Websites')
ax.set_ylabel('Ratings')
ax.set_title('Different website ratings')
plt.show()
```

```python
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

fig, ax = plt.subplots()
position = arange(5)+ 0.75
height = ratings.loc[:, num_cols ].iloc[0, :]
width = 0.5
ax.barh(position, height, width)
ax.set_yticks(arange(5)+ 0.75)
ax.set_yticklabels(num_cols)
ax.set_ylabel('Websites')
ax.set_xlabel('Ratings')
ax.set_title('Different website ratings')
plt.show()
```

## Scatter plots 

Comparing Fandango to Metacritic, IMDB and Rotten Tomatoes.

```python
fig = plt.figure(figsize= (10,20))
ax1 = fig.add_subplot(3, 1, 1)
ax2 = fig.add_subplot(3, 1, 2)
ax3 = fig.add_subplot(3, 1, 3)

ax1.scatter(ratings.loc[:, 'Fandango_Ratingvalue'], ratings.loc[:, 'RT_user_norm'])
ax1.set_xlabel('Fandango')
ax1.set_ylabel('Rotten Tomatoes')
ax1.set_xlim(0, 5)
ax1.set_ylim(0, 5)

ax2.scatter(ratings['Fandango_Ratingvalue'], ratings['Metacritic_user_nom'])
ax2.set_xlabel('Fandango')
ax2.set_ylabel('Metacritic')
ax2.set_xlim(0, 5)
ax2.set_ylim(0, 5)

ax3.scatter(ratings['Fandango_Ratingvalue'], ratings['IMDB_norm'])
ax3.set_xlabel('Fandango')
ax3.set_ylabel('IMDB')
ax3.set_xlim(0, 5)
ax3.set_ylim(0, 5)

plt.title('Scatter plot for Different websites vs Fandango')
plt.show()
```

## Histograms

```python
fig, ax= plt.subplots()
ax.hist(ratings['Fandango_Ratingvalue'], range= (0,5))
plt.show()
```

```python
fig = plt.figure(figsize=(10,20))
ax1 = fig.add_subplot(4,1,1)
ax2 = fig.add_subplot(4,1,2)
ax3 = fig.add_subplot(4,1,3)
ax4 = fig.add_subplot(4,1,4)

ax1.hist(ratings['Fandango_Ratingvalue'], bins=20, range=(0, 5))
ax1.set_title('Distribution of Fandango Ratings')
ax1.set_ylim(0, 50)

ax2.hist(ratings['RT_user_norm'], 20, range=(0, 5))
ax2.set_title('Distribution of Rotten Tomatoes Ratings')
ax2.set_ylim(0, 50)

ax3.hist(ratings['Metacritic_user_nom'], 20, range=(0, 5))
ax3.set_title('Distribution of Metacritic Ratings')
ax3.set_ylim(0, 50)

ax4.hist(ratings['IMDB_norm'], 20, range=(0, 5))
ax4.set_title('Distribution of IMDB Ratings')
ax4.set_ylim(0, 50)

plt.show()
```

```python
fig, ax = plt.subplots()
ax.boxplot(ratings['RT_user_norm'])
ax.set_xticklabels(['Rotten Tomatoes'])
ax.set_ylim(0, 5)
plt.show()
```

```python
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue']
fig, ax = plt.subplots()
ax.boxplot(ratings[num_cols].values)
ax.set_xticklabels(num_cols, rotation=90)
ax.set_ylim(0,5)
plt.show()
```
